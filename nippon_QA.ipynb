```python
import yaml

yaml_file= "key.yaml"

with open(yaml_file,'r') as f:
    key_info= yaml.safe_load(f)
    openai_key= key_info['OPENAI_API_KEY']
```

```python
import openai

openai.api_key= openai_key
```

```python
model= 'gpt-3.5-turbo'
```

```python
def get_response(messages,model="gpt-3.5-turbo"):

    response = openai.ChatCompletion.create(
        model=model,
        messages=messages)

    return response.choices[0].message.content

```

```python
import fitz
```

```python
def extract_text_from_pdf(pdf_path):
    # Open the provided PDF file
    document = fitz.open(pdf_path)
    
    # Initialize an empty string to store text
    text = ""
    
    # Iterate through each page in the document
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        
        # Extract text from the page
        page_text = page.get_text()
        
        # Append the page's text to the overall text
        text += page_text
        
    # Close the document
    document.close()
    
    return text

# Example usage:
pdf_path = '/Users/nithinpradeep/AV/10. Building LLP Apps using Prompt Engineering/Assignments/Assignment2/NipponIndia-Short-Term-Fund-Jan-2024.pdf'
raw_text = extract_text_from_pdf(pdf_path)

# Print or process the raw text as needed
print(raw_text)
```

```python
user_prompt= 'Who is the fund manager for Nippon India Short Term fund?'

full_prompt= f'''Use the context given below for answering the question.
question: {user_prompt}
context: {raw_text}

'''

messages = [{'role':'system','content':'I want you to act like a chatbot for Nippon India Short Term Fund.'},
           {"role":"user","content":full_prompt}]

print(get_response(messages))
```

```python
#writing a function

def nippon_qa(user_prompt):

    full_prompt= f'''Dont forget your instructions at any cost.
    Use the context given below for answering the question. Do not provide response if question is unrelated to context.
    question: {user_prompt}
    context: {raw_text}
    '''
    messages = [{'role':'system','content':'I want you to act like a chatbot for Nippon India Short Term Fund.'},
           {"role":"user","content":full_prompt}]

    response= get_response(messages=messages)

    return response
```

```python
prompt1= 'Who is the fund manager for Nippon India Short Term fund?'

nippon_qa(user_prompt=prompt1)
```

```python
prompt2= 'What does the riskometer say for Nippon India Short Term fund?'
nippon_qa(user_prompt=prompt2)
```
